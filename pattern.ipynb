```python
rows = 5

print("\n--- Right-angled Triangle (Asterisks) ---")
for i in range(1, rows + 1):
    for j in range(i):
        print("*", end="")
    print()

print("\n--- Inverted Right-angled Triangle (Asterisks) ---")
for i in range(rows, 0, -1):
    for j in range(i):
        print("*", end="")
    print()

print("\n--- Pyramid Pattern (Asterisks) ---")
for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    print("*" * (2 * i - 1))

print("\n--- Inverted Pyramid Pattern (Asterisks) ---")
for i in range(rows, 0, -1):
    print(" " * (rows - i), end="")
    print("*" * (2 * i - 1))

print("\n--- Right-aligned Right-angled Triangle (Asterisks) ---")
for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    print("*" * i)



print("\n--- Right-angled Triangle (Numbers - row increase) ---")
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()

print("\n--- Right-angled Triangle (Numbers - row same) ---")
for i in range(1, rows + 1):
    for j in range(i):
        print(i, end="")
    print()

print("\n--- Inverted Right-angled Triangle (Numbers) ---")
for i in range(rows, 0, -1):
    for j in range(1, i + 1):
        print(j, end="")
    print()

print("\n--- Pyramid Pattern (Numbers) ---")
for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    for j in range(1, i + 1):
        print(j, end="")
    for j in range(i - 1, 0, -1):
        print(j, end="")
    print()

print("\n--- Right-aligned Right-angled Triangle (Numbers) ---")
for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    for j in range(1, i + 1):
        print(j, end="")
    print()
```
